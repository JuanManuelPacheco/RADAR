### Fourier analysis of the radio signals

```python
Spx = np.fft.fft(Ey)
Spxg = np.fft.fft(Eyg)
Spxgw = np.fft.fft(Eygw)
Spxw = np.fft.fft(Eyw)
timestep = 2e-10  # Time resolution observer from SIM000001.reas file

# Find the positive frequencies
freq = np.fft.fftfreq( Spx.size, d = timestep )
index = np.where(freq >= 0.)

freqg = np.fft.fftfreq( Spxg.size, d = timestep )
indexg = np.where(freqg >= 0.)

freqgw = np.fft.fftfreq( Spxgw.size, d = timestep )
indexgw = np.where(freqgw >= 0.)

freqw = np.fft.fftfreq( Spxw.size, d = timestep )
indexw = np.where(freqw >= 0.)

# Scale the real part of the FFT and clip the data
clipped_spectrum = timestep*Spx[index].real
clipped_frequency = freq[index]

clipped_spectrumg = timestep*Spxg[index].real
clipped_frequencyg = freqg[index]

clipped_spectrumgw = timestep*Spxgw[index].real
clipped_frequencygw = freqgw[index]

clipped_spectrumw = timestep*Spxw[index].real
clipped_frequencyw = freqw[index]
```

```python
fig = plt.figure(figsize=(10, 5))
plt.plot(clipped_frequency/1e6, np.abs(clipped_spectrum), 'b')
#plt.plot(clipped_frequencyg/1e6, np.abs(clipped_spectrumg), 'r')
plt.plot(clipped_frequencygw/1e6, np.abs(clipped_spectrumgw), 'r')
#plt.plot(clipped_frequencyw/1e6, np.abs(clipped_spectrumw), 'g')

plt.xlabel("Frequency [MHz]", fontsize = 15)
plt.ylabel("Power [V/ m * MHz]", fontsize = 15)
#plt.legend(['Ey','Ey_GDAS','Ey_GDAS_WMM','Ey_WMM'], loc=0)
plt.legend(['Ey','Ey_GDAS_WMM'], loc=0)
plt.title("Frecuencias en la antena (1) para un $p^+$ de $10^{17}$ eV (Berlin)", fontsize = 15)
plt.axis([0, 300, 0, 2.7e-18])
plt.grid()
```

### Array setup

```python
x_pos = []
y_pos = []

with open("SIM000001.list") as f:
    lines = f.readlines()
    for line in lines:
        if (line.find("AntennaPosition")== 0):  # Searching for antenna positions
            lin = line.split(" ")
            pos = lin[2].split('\t')
            x_pos.append(pos[0])
            y_pos.append(pos[1])
```

```python
x_pos = np.array(x_pos)
y_pos = np.array(y_pos)

fig = plt.figure(figsize=(5, 5))
plt.plot(x_pos, y_pos, '*r', markersize=20)
plt.xlabel("X [cm]", fontsize = 20)
plt.ylabel("Y [cm]", fontsize = 20)
plt.axis([-10000, 50000, 10000, -50000])
plt.grid()
```
